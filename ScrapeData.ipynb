# Looking at the HTML
First let's take a at the HTML on the web:

For a given vote, breakdown by senator
 * http://www.senate.gov/legislative/LIS/roll_call_lists/roll_call_vote_cfm.cfm?congress=110&session=2&vote=00035
 
For all votes (no breakdown)

```python
# Download the raw HTML
import urllib2
url = "http://www.senate.gov/legislative/LIS/roll_call_lists/roll_call_vote_cfm.cfm?congress=110&session=2&vote=00035"
vote_35_html = urllib2.urlopen(url).read()
```

```python
print(vote_35_html)
```

```python
lines = vote_35_html.splitlines()
```

```python
print(lines[222])
```

```python
# Let's try searching for the Yeas.
# Maybe we can do something easy, like "if a senator's name
# is on a line, and so is "Yea", mark them as "Yea".
for line in lines:
    if line.find('Yea') != -1:
        print(line)
```

The listing by state screws that up. But there are tools that will help here...
* https://xkcd.com/1171/
* https://docs.python.org/2/library/re.html

```python
# Let's create a regular expression to match the data that we
# want to find. Let's start with just finding
# Name (X-XX)
import re
my_regular_expression  = '\w+ \(.-..\)'
all_found = re.findall(my_regular_expression, vote_35_html)
print(all_found)
```

```python
# We can print this line-by-line as
for found in all_found:
    print(found)
```

```python
# There's a bug here... this can get painful
# Can you find Lieberman?
```

```python
# Fix that bug by allowing 1 or 2 characters for party
# The way to say 1 character is "."
# The way to say 1 or more characters is ".+"
# The way to say 1 or 2 characters is ".{1,2}"
my_regular_expression  = '\w+ \(.-..\)'
my_regular_expression  = '\w+ \(.{1,2}-..\)'
my_regular_expression = '\w+ \(\w+-..\)'
all_found = re.findall(my_regular_expression, vote_35_html)
for found in all_found:
    print(found)
```

```python
# Remember how parentheses were weird?
# We can pull out just the senator name by putting it in ()
# We started with
#   "\w+ \(.-..\)"
# change this to
#   "(\w+) \(.-..\)"
my_regular_expression  = '\w+ \(.{1,2}-..\)'
my_regular_expression  = '(\w+) \(.{1,2}-..\)'
all_found = re.findall(my_regular_expression, vote_35_html)
for found in all_found:
    print(found)
```

```python
# Let's put the party and state and () as well...
my_regular_expression = '(\w+) \(.{1,2}-..\)'
my_regular_expression = '(\w+) \((.{1,2})-(..)\)'
all_found = re.findall(my_regular_expression, vote_35_html)
for found in all_found:
    print(found)
```

```python
# And ... let's expand this to include the vote
my_regular_expression = '(\w+) \((.{1,2})-(..)\), <b>(Yea|Nay|Not Voting)</b>'
all_found = re.findall(my_regular_expression, vote_35_html)
for found in all_found:
    print(found)
```

```python
# Let's start importing this into dictionary
senators = set()
senator_party = {}
senator_state = {}
vote_35 = {}
for found in all_found:
    senator = found[0]
    party = found[1]
    state = found[2]
    vote = found[3]
    senators.add(senator)
    vote_35[senator] = vote
    senator_party[senator] = party
    senator_state[senator] = state
print("Number of voting senators: %d" % len(senators))
```
