```python
# Clear our throat
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import random
%matplotlib inline
sns.set_style("white")
```

```python
# Make some noise!
x_noise = np.concatenate((100 * np.random.randn(100),  50 * np.random.randn(100)), 0)
y_noise = np.concatenate((200 * np.random.randn(100), 100 * np.random.randn(100)), 0)

# Make some signal!
x_signal = 50 * np.random.randn(100)
y_signal = 2 * x_signal + 20
```

```python
# Show them
fig = plt.figure()
plt.figure(fig.number)
plt.scatter(x_noise, y_noise, label='Noise', color='blue')
plt.scatter(x_signal, y_signal, label='Signal', color='red')
plt.legend()
plt.show()
```

```python
fig_dist = plt.figure()
x = np.concatenate((x_signal, x_noise), 0)
y = np.concatenate((y_signal, y_noise), 0)
```

```python
i = random.sample(range(len(x)),2)

poly = np.polyfit(x[i], y[i], 1)
y_fit = np.polyval(poly, x)

plt.figure(fig.number)
fig.clear()
fig.canvas.draw()
plt.scatter(x, y, color='blue', label='Samples')
plt.plot(x, y_fit, color='red', label='Fit')
plt.legend()
fig.canvas.draw()
```

```python
err = np.sqrt(np.power(y - y_fit, 2))
plt.figure(fig_dist.number)
fig_dist.clear()
sns.distplot(err)
fig_dist.canvas.draw()
total_err = sum(err)
quarter_err = np.sort(err)[int(len(err)/4)]
inliner_epsilon = 1
inliers = sum(err < inliner_epsilon)
print 'Error: %f, Error at 25th percentile: %f, %d inliers' % (total_err, quarter_err, inliers)
```
